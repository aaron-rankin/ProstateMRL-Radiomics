```python
url = 'D:/data/prostateMR_radiomics/nifti/20fractions/'
ptDir = os.listdir(url)

print(ptDir)
```

```python
def MeanStdExtraction(imageArray, mask_url):
    mask = sitk.ReadImage(mask_url)
    maskArray = sitk.GetArrayFromImage(mask)

    print(maskArray.max())
    print(imageArray.shape)

    maskedImage = ma.MaskedArray(imageArray, mask = np.logical_not(maskArray),keep_mask = True, hard_mask = True)
    maskedImage = maskedImage.flatten()
    
    imageMean = np.mean(maskedImage)
    imageStd = np.std(maskedImage)

    return imageMean, imageStd
```

```python
for i in ptDir:
    scanWeeks = os.listdir(url + str(i))

    print(i)

    for j in scanWeeks:
        NiiFiles = os.listdir(url + str(i) + "/" + str(j))
        print ("Processing: "+i+"  Timepoint: "+j)

        image = url+str(i)+"\\"+str(j)+"\\"+str(i)+"_"+str(j)+"_image.nii"
        bodyMask = url+str(i)+"\\"+str(j)+"\\"+str(i)+"_"+str(j)+"_body_mask.nii"
        
        readImage = sitk.ReadImage(image)
        readBodyMask = readBodyMask = sitk.ReadImage(bodyMask)

        bodyMaskArray = sitk.GetArrayFromImage(readBodyMask)
        imageArray = sitk.GetArrayFromImage(readImage)

        # remove stray pixel values 
        imageArray = imageArray * bodyMaskArray


        for k in NiiFiles:

            if "psoas" in k:
                print(k)

                mask_url = url + str(i) + "\\" + str(j) + "\\" + str(k)

                mean = MeanStdExtraction(imageArray, mask_url)[0]

                print("image Mean: " + str(mean))


```
